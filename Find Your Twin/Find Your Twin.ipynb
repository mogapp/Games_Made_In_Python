# Find Your Twin

```python
import random
import datetime

birthday=[]
i=0
while(i<50):
        
    yr=random.randint(1895,2019)
        
    if (yr%4==0 and yr%100!=0 and yr%400==0):
        leapyr=1
            
    else:
        leapyr=0
            
    mon=random.randint(1,12)
    if (mon==2 and leapyr==1):
        day=random.randint(1,29)
    elif (mon==2 and leapyr==0):
        day=random.randint(1,28)
    elif (mon%2!=0 and mon<7):
        day=random.randint(1,31)
    elif (mon%2==0 and mon>7 and mon<12):
        day=random.randint(1,31)
    elif(mon==7 or mon==8):
        day=random.randint(1,31)
    else:
        day=random.randint(1,30)

    dd=datetime.date(yr,mon,day)
    day_of_year=dd.timetuple().tm_yday
    birthday.append(day_of_year)
    i=i+1

    


birthday.sort()
i=0
twin=0
while(i<50):
    print(birthday[i])
    try:
        if (birthday[i]==birthday[i+1] or birthday[i-1]==birthday[i]):
            twin=twin+1
    except:
        pass
    i=i+1
print("TWINS ARE: ",int(twin/2))
```
