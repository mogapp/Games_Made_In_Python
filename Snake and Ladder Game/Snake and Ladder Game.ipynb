# Snake and Ladder Game

```python
import random

def check_snake(n):
    if n==44:
        return 19
    elif n==46:
        return 5
    elif n==48:
        return 9
    elif n==52:
        return 11
    elif n==55:
        return 7
    elif n==59:
        return 17
    elif n==64:
        return 36
    elif n==69:
        return 33
    elif n==73:
        return 1
    elif n==83:
        return 19
    elif n==92:
        return 51
    elif n==95:
        return 24
    elif n==98:
        return 24
    else:
        return n
    

def check_ladder(n):
    if n==8:
        return 26
    elif n==21:
        return 82
    elif n==43:
        return 77
    elif n==50:
        return 91
    elif n==62:
        return 96
    elif n==66:
        return 87
    elif n==54:
        return 93
    elif n==80:
        return 100
    else:
        return n

def play():
    turn=0
    pp1=0
    pp2=0
    p1name=input("Enter the name of player 1: ")
    p2name=input("Enter the name of player 2: ")
    while(1):
        if(turn%2):
            print("\n",p1name, ", Its your turn")
            c=input("Press 1 to continue and 0 to exit: ")
            if(c=='0'):
                print(p1name,", your score is: ",pp1)
                print(p2name,", your score is: ",pp2)
                if(pp1>=100):
                    print(p1name," you won")
                elif pp2>=100:
                    print(p2name, " you won")
                print("Thanks for playing.")
                break
            else:
                dice=random.randint(1,6)
                print("Dice has scored: ",dice)
                pp1=pp1+dice
                pp1=check_ladder(pp1)
                pp1=check_snake(pp1)
                print("Your point is ",pp1)
                if pp1>=100:
                    print(p1name,", You won the game ")
                    break
        else:
            print("\n",p2name, ", Its your turn")
            c=input("Press 1 to continue and 0 to exit: ")
            if(c=='0'):
                print(p2name,", your score is: ",pp2)
                print(p1name,", your score is: ",pp1)
                if(pp1>=100):
                    print(p1name," you won")
                elif pp2>=100:
                    print(p2name, " you won")
                print("Thanks for playing.")
                break
            else:
                dice=random.randint(1,6)
                print("Dice has scored: ",dice)
                pp2=pp2+dice
                pp2=check_ladder(pp2)
                pp2=check_snake(pp2)
                print("Your point is ",pp2)
                if pp2>=100:
                    print(p2name,", You won the game ")
                    break

        turn=turn+1


play()
```
